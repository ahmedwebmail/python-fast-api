```python
import requests
```

```python
url = 'https://jsonplaceholder.typicode.com/posts/'
res = requests.get(url)

if res.status_code == 200:
  print(res.json()[0])
else:
  print('Error')
```

```python
url = 'https://jsonplaceholder.typicode.com/posts/'

data = {
    'title': 'sunt aut facere repellat ahmed',
    'body': 'quia et suscipit\nsuscipit recusandae consequuntur expedita et cum\nreprehenderit molestiae ut ut quas totam\nnostrum rerum est ahmed',
    'userId': 1
}

res = requests.post(url, json=data)

if res.status_code == 201:
  created_post = res.json()
  print(created_post)
else:
  print("Faild")
```
